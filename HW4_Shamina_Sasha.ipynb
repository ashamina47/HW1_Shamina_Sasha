QUESTION 5

```python
import matplotlib.pyplot as plt
import numpy as np
marks = {'Andy':88, 'Amy':66, 'James': 90, 'Jules': 55, 'Arthur': 77}
keys = marks.keys()
values = marks.values()

plt.bar(keys, values)
plt.xlabel("Students")
plt.ylabel("Grades")
plt.title("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
marks = {'Andy':88, 'Amy':66, 'James': 90, 'Jules': 55, 'Arthur': 77}
keys = marks.keys()
values = marks.values()

plt.pie(values, labels = keys)
plt.legend(title = "Marks")
plt.show()
```
