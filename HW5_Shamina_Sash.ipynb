```r
Data_Frame <- data.frame(
id = c(1, 2, 3, 4, 5),
name = c("Peter", "Amy", "Ryan", "Gary", "Michelle"),
salary = c(623.30, 515.20, 611.00, 729.00, 843.25)
)
Data_Frame
```

b.

```r
New_DF <- cbind(Data_Frame, department = c("IT", "finance", "computer science", "HR", "management"))
New_DF
```

c.

```r
New_DF[c(1,3,5),c(2,3)]
```

d.

```r
Bar_DF = New_DF[c(1,4,5), ]
Bar_DF
```

```r
with(Bar_DF, barplot(salary, names.arg = name, col = "purple", main = "names and salaries"))
```

e.

```r
highest <- max(Data_Frame$salary)
lowest <- min(Data_Frame$salary)
median <- median(Data_Frame$salary)
Pie_DF <- data.frame(
StatName = c( "Highest Salary", "Lowest Salary", "Median Salary"),
StatNumbers = c(highest, lowest, median))
colors <- c("slateblue1", "skyblue3", "palegreen3")
Pie_DF
```

```r
pie(Pie_DF$StatNumbers, label = Pie_DF$StatName, main = "salary stats", col = colors) 
legend("topleft", title="salary stats", legend = Pie_DF$StatName, fill = colors)
```

Question 3

```r
amazon = read.csv('report.csv')
head(amazon)
```

```r
Amazon <- amazon[c("Order.Date", "Category", "Item.Subtotal.Tax", "Item.Total")]
Amazon
```

```r
Amazon$Item.Total = as.numeric(gsub("\\$", "", Amazon$Item.Total)) 
Amazon$Item.Subtotal.Tax = as.numeric(gsub("\\$", "", Amazon$Item.Subtotal.Tax))
Amazon$Item.Total
Amazon$Item.Subtotal.Tax
```

```r
summary(Amazon$Item.Total)
summary(Amazon$Item.Subtotal.Tax)
```

```r
barplot(Amazon$Item.Total, names.arg = Amazon$Order.Date, main = "Total Spending", col = "skyblue4")
barplot(Amazon$Item.Subtotal.Tax, names.arg = Amazon$Order.Date, main = "Total Tax Spending", col = "sienna3")
```

```r
pie(Amazon$Item.Total, label = Amazon$Category, cex=0.3, radius = 0.5, main = "Spending Based on Category")
legend("topleft", title="Categoies", legend = Amazon$Category, cex = 0.3)
pie(Amazon$Item.Subtotal.Tax, label = Amazon$Category, cex=0.3, radius = 0.5, main = " Tax Spending Based on Category")
legend("topleft", title="Categoies", legend = Amazon$Category, cex = 0.3)
```

```r
plot(Amazon$Order.Date, Amazon$Item.Total, col = 'blue')
```

```r
plot(Amazon$Item.Total, type = 'solid', col = "salmon2")
lines(Amazon$Item.Subtotal.Tax, type = 'solid', col = "royalblue2")
```
