```python
import pandas as pd
df = pd.read_csv('report.csv')
df.head()
```

```python
df.shape
```

```python
df = df.fillna(0)
df.head()
```

```python
df["Item Total"] = df["Item Total"].astype(float)
df.head()
```

```python
df["Item Total"].sum()
```

```python
df["Item Total"].mean()
```

```python
df["Item Total"].median()
```

```python
df["Item Total"].max()
```

```python
df["Item Total"].min()
```

```python
df["Item Total"].std()
```

```python
df["Item Subtotal Tax"] = df["Item Subtotal Tax"].str.replace('$','').astype(float)
df.head()
```

```python
df["Item Subtotal Tax"].sum()
```

```python
df["Item Subtotal Tax"].mean()
```

```python
df["Item Subtotal Tax"].median()
```

```python
df["Item Subtotal Tax"].max()
```

```python
df["Item Subtotal Tax"].min()
```

```python
df["Item Subtotal Tax"].std()
```

```python
df.plot.bar(x="Order Date", y="Item Total", color = 'royalblue')
```

```python
df.plot.bar(x="Order Date", y="Item Subtotal Tax", color = 'crimson')
```

```python
df["Category"]
```

```python
daily_orders = df.groupby("Category").sum()["Item Total"]
daily_orders.head()
```

```python
daily_orders.plot.pie(x = "Category", y = "Item Total", figsize=(10,15),labeldistance=None).legend(loc='upper right')
```

```python
daily_orders = df.groupby("Category").sum()["Item Subtotal Tax"]
daily_orders.head()
```

```python
daily_orders.plot.pie(x = "Category", y = "Item Subtotal Tax", figsize=(10,15),labeldistance=None).legend(loc='upper right')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
plt.plot(df["Order Date"], df["Item Total"])
plt.plot(df["Order Date"], df["Item Subtotal Tax"])

plt.show()
```
